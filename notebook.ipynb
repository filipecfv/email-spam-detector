# Importing Data
## Detecting email spam based on word frequencies

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('input/spambase.data', header = None)
df 
```

This dataset originally has no column names. Each column or attribute represents a certain word whose frequent has been measured. 

The last column is a dummy variable (0/1) that indicates whether that email is spam or not. 

(Read "input/spambase.DOCUMENTATION" for more information.)

```python
# to check whether the last column is really a dummy variable: 
df[57].unique()
```

```python
# let us rename the last attribute (57) as spam
df = df.rename(columns = {57:'spam'})

df.to_csv('output/spam_email.csv') # writing
df
```

```python
df.describe()
# general stats
```

```python
# correlation matrix between each attribute and the spam dummy 

correlations = df.corr()
spam_correlation = correlations['spam'].to_frame()
spam_correlation = spam_correlation.sort_values('spam', ascending = False)
spam_correlation = spam_correlation.transpose()

spam_correlation.to_csv('output/spam-attribute_correlation.csv')
spam_correlation
```

# Machine Learning modeling
## Logistic Regression

```python
from sklearn.linear_model import LogisticRegression # scikit learn
from sklearn.model_selection import train_test_split
```

```python
# spliting test and train samples
df_predictors = df.drop('spam', axis = 1)
df_predicted = df['spam']

X_train, X_test, y_train, y_test = train_test_split(df_predictors, df_predicted)
```

```python
# training the model 
LRmodel = LogisticRegression(solver='lbfgs', max_iter=1600)
LRmodel.fit(X_train, y_train)
```

```python
# prediction score
LRmodel.score(X_test, y_test)
```

The model is capable to detect in average 92.34% of spam emails;
